```python
import os
import itertools
from PeptideBuilder import PeptideBuilder
from Bio.PDB import PDBIO
```

```python
path = r'/data4/liangdh/project/alnp/task/pdb'
```

```python
phi = [-57.0, -57.0]
psi = [-47.0, -47.0]
omega = [180.0, 180.0]
aas = list('GAVLIMPWSYCFNQTDEKRH')
tripeptides = itertools.product(aas, repeat=3)
tripeptide_list = [''.join(triplet) for triplet in tripeptides]
```

```python
for seq in tripeptide_list:
    structure = PeptideBuilder.make_structure(seq, phi, psi, omega)
    io = PDBIO()
    io.set_structure(structure)
    io.save(os.path.join(path,f"{seq}.pdb"))
```
